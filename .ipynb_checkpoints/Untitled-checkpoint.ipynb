```python
import os

def contains_target_files(directory, extensions):
    """
    Check if a directory contains files with the given extensions
    """
    for item in os.listdir(directory):
        if os.path.isdir(os.path.join(directory, item)):
            # Recursively check in subdirectories
            if contains_target_files(os.path.join(directory, item), extensions):
                return True
        elif any(item.endswith(ext) for ext in extensions):
            return True
    return False

def print_directory_hierarchy(path, indent=0, extensions=('.py', '.html')):
    # Check if the path is a directory
    if os.path.isdir(path):
        # Only proceed if the directory contains target files
        if contains_target_files(path, extensions):
            # Print the directory name with a trailing slash
            print('    ' * indent + os.path.basename(path) + '/')
            # Iterate over all items in the directory
            for item in os.listdir(path):
                # Recursively call this function for each item
                print_directory_hierarchy(os.path.join(path, item), indent + 1, extensions)
    else:
        # If it's a file, print its name only if it has the desired extension
        if path.endswith(extensions):
            print('    ' * indent + os.path.basename(path))

# Example usage
print_directory_hierarchy('/Users/mm/Desktop/courier_tracking')



```

```python
import os
import graphviz

def visualize_directory_graphviz(directory_path):
    """Generates a directory tree visualization using Graphviz.

    Args:
        directory_path: The path to the directory to visualize.
    """

    dot = graphviz.Digraph(comment='Directory Structure')

    def add_nodes(path, parent_node):
        for item in os.listdir(path):
            full_path = os.path.join(path, item)            
            if os.path.isdir(full_path):
                node_name = item  
                dot.node(node_name, item)  # Create a node for the directory
                dot.edge(parent_node, node_name)  # Link to parent
                add_nodes(full_path, node_name)  # Recurse into subdirectories
            else:
                dot.node(item, item)  # Create a node for a file
                dot.edge(parent_node, item)

    add_nodes(directory_path, '')  # Start recursion from the root directory

    dot.render(view=True)  # Render and view the graph (requires Graphviz installed)

# Example usage:
visualize_directory_graphviz('/Users/mm/Desktop/courier_tracking')
```

```python
import random
import time
import tkinter as tk
from threading import Thread

WIDTH = 1200
HEIGHT = 900
DELAY = 10

array = [random.randint(1, 100) for _ in range(40)]
quickSortArray = array.copy()
mergeSortArray = array.copy()
heapSortArray = array.copy()
insertionSortArray = array.copy()
radixSortArray = array.copy()
selectionSortArray = array.copy()

algorithmNames = [
    "QuickSort", "MergeSort", "HeapSort",
    "InsertionSort", "RadixSort", "SelectionSort"
]

max_value = max(array)

class SortingComponent:
    def __init__(self, visualizer, algorithm, array, index):
        self.visualizer = visualizer
        self.algorithm = algorithm
        self.array = array
        self.index = index
        self.start_time = None
        self.completed = False

    def run(self):
        self.start_time = time.time()
        if self.algorithm == "quick_sort":
            self.quick_sort(0, len(self.array) - 1)
        elif self.algorithm == "merge_sort":
            self.merge_sort(0, len(self.array) - 1)
        elif self.algorithm == "heap_sort":
            self.heap_sort()
        elif self.algorithm == "insertion_sort":
            self.insertion_sort()
        elif self.algorithm == "radix_sort":
            self.radix_sort()
        elif self.algorithm == "selection_sort":
            self.selection_sort()
        self.completed = True

    def quick_sort(self, low, high):
        if low < high:
            pivot_index = self.partition(low, high)
            self.quick_sort(low, pivot_index - 1)
            self.quick_sort(pivot_index + 1, high)

    def partition(self, low, high):
        pivot = self.array[high]
        i = low - 1
        for j in range(low, high):
            if self.array[j] <= pivot:
                i += 1
                self.swap(i, j)
                self.visualizer.update_bars(self.index, self.array)
                self.visualizer.update()
                time.sleep(DELAY / 1000)
        self.swap(i + 1, high)
        self.visualizer.update_bars(self.index, self.array)
        self.visualizer.update()
        time.sleep(DELAY / 1000)
        return i + 1

    def merge_sort(self, left, right):
        if left < right:
            mid = left + (right - left) // 2
            self.merge_sort(left, mid)
            self.merge_sort(mid + 1, right)
            self.merge(left, mid, right)

    def merge(self, left, mid, right):
        temp = [0] * (right - left + 1)
        i, j, k = left, mid + 1, 0

        while i <= mid and j <= right:
            if self.array[i] <= self.array[j]:
                temp[k] = self.array[i]
                i += 1
            else:
                temp[k] = self.array[j]
                j += 1
            k += 1

        while i <= mid:
            temp[k] = self.array[i]
            i += 1
            k += 1

        while j <= right:
            temp[k] = self.array[j]
            j += 1
            k += 1

        for l in range(len(temp)):
            self.array[left + l] = temp[l]
            self.visualizer.highlight_bar(self.index, left + l)
            self.visualizer.update_bars(self.index, self.array)
            self.visualizer.update()
            time.sleep(DELAY / 1000)

    def heap_sort(self):
        n = len(self.array)

        for i in range(n // 2 - 1, -1, -1):
            self.heapify(n, i)

        for i in range(n - 1, 0, -1):
            self.visualizer.highlight_bar(self.index, 0)
            self.swap(0, i)
            self.visualizer.update_bars(self.index, self.array)
            self.visualizer.update()
            time.sleep(DELAY / 1000)
            self.heapify(i, 0)

    def heapify(self, n, i):
        largest = i
        left = 2 * i + 1
        right = 2 * i + 2

        if left < n and self.array[left] > self.array[largest]:
            largest = left

        if right < n and self.array[right] > self.array[largest]:
            largest = right

        if largest != i:
            self.swap(i, largest)
            self.visualizer.update_bars(self.index, self.array)
            self.visualizer.update()
            time.sleep(DELAY / 1000)
            self.heapify(n, largest)

    def insertion_sort(self):
        for i in range(1, len(self.array)):
            key = self.array[i]
            j = i - 1
            while j >= 0 and self.array[j] > key:
                self.array[j + 1] = self.array[j]
                j -= 1
                self.visualizer.highlight_bar(self.index, j + 1)
                self.visualizer.update_bars(self.index, self.array)
                self.visualizer.update()
                time.sleep(DELAY / 1000)
            self.array[j + 1] = key
            self.visualizer.highlight_bar(self.index, i)
            self.visualizer.update_bars(self.index, self.array)
            self.visualizer.update()
            time.sleep(DELAY / 1000)

    def radix_sort(self):
        max_val = max(self.array)
        exp = 1
        while max_val // exp > 0:
            self.count_sort(exp)
            exp *= 10

    def count_sort(self, exp):
        n = len(self.array)
        output = [0] * n
        count = [0] * 10

        for i in range(n):
            count[(self.array[i] // exp) % 10] += 1

        for i in range(1, 10):
            count[i] += count[i - 1]

        i = n - 1
        while i >= 0:
            output[count[(self.array[i] // exp) % 10] - 1] = self.array[i]
            count[(self.array[i] // exp) % 10] -= 1
            i -= 1

        for i in range(n):
            self.array[i] = output[i]
            self.visualizer.highlight_bar(self.index, i)
            self.visualizer.update_bars(self.index, self.array)
            self.visualizer.update()
            time.sleep(DELAY / 1000)

    def selection_sort(self):
        for i in range(len(self.array) - 1):
            min_index = i
            for j in range(i + 1, len(self.array)):
                if self.array[j] < self.array[min_index]:
                    min_index = j
                self.visualizer.highlight_bar(self.index, j)
                self.visualizer.update_bars(self.index, self.array)
                self.visualizer.update()
                time.sleep(DELAY / 1000)
            self.swap(i, min_index)
            self.visualizer.update_bars(self.index, self.array)
            self.visualizer.update()
            time.sleep(DELAY / 1000)

    def swap(self, i, j):
        self.array[i], self.array[j] = self.array[j], self.array[i]
        self.visualizer.highlight_bar(self.index, i)
        self.visualizer.highlight_bar(self.index, j)

class ArrayVisualizer(tk.Tk):
    def __init__(self):
        super().__init__()
        self.title("Array Visualization")
        self.geometry(f"{WIDTH}x{HEIGHT}")
        self.canvas = tk.Canvas(self, width=WIDTH, height=HEIGHT, bg="black")
        self.canvas.pack()

        self.bars = [[] for _ in range(6)]
        self.labels = [None] * 6
        self.algorithm_texts = [None] * 6

        self.create_panes()

        self.sorting_components = [
            SortingComponent(self, "quick_sort", quickSortArray, 0),
            SortingComponent(self, "merge_sort", mergeSortArray, 1),
            SortingComponent(self, "heap_sort", heapSortArray, 2),
            SortingComponent(self, "insertion_sort", insertionSortArray, 3),
            SortingComponent(self, "radix_sort", radixSortArray, 4),
            SortingComponent(self, "selection_sort", selectionSortArray, 5)
        ]

        self.start_sorting_algorithms()
        self.update_animation()

    def create_panes(self):
        pane_width = (WIDTH - 40) // 3
        pane_height = (HEIGHT - 160) // 2

        for i in range(6):
            pane_x = (i % 3) * (pane_width + 20) + 10
            pane_y = (i // 3) * (pane_height + 80) + 60

            for j in range(len(array)):
                bar_width = pane_width / len(array)
                bar_height = array[j] * pane_height / max_value
                bar = self.canvas.create_rectangle(
                    pane_x + j * bar_width, pane_y + pane_height - bar_height,
                    pane_x + (j + 1) * bar_width, pane_y + pane_height,
                    fill="white", outline="")
                self.bars[i].append(bar)

            label = self.canvas.create_text(
                pane_x + pane_width / 2, pane_y + pane_height + 20, text="Time elapsed: 0 ms",
                fill="white", font=("Arial", 12))
            self.labels[i] = label

            algorithm_text = self.canvas.create_text(
                pane_x + pane_width / 2, pane_y + 20, text=algorithmNames[i],
                fill="white", font=("Arial", 16, "bold"))
            self.algorithm_texts[i] = algorithm_text

    def update_animation(self):
        all_completed = True
        for i in range(6):
            if self.sorting_components[i].start_time is not None:
                self.update_time_label(i, self.sorting_components[i].start_time)
                if not self.sorting_components[i].completed:
                    all_completed = False
        if not all_completed:
            self.after(DELAY, self.update_animation)

    def update_bars(self, index, arr):
        pane_width = (WIDTH - 40) // 3
        pane_height = (HEIGHT - 160) // 2
        pane_x = (index % 3) * (pane_width + 20) + 10
        pane_y = (index // 3) * (pane_height + 80) + 60

        for i in range(len(arr)):
            bar_height = arr[i] * pane_height // max_value
            self.canvas.coords(
                self.bars[index][i],
                pane_x + i * pane_width // len(arr), pane_y + pane_height - bar_height,
                pane_x + (i + 1) * pane_width // len(arr), pane_y + pane_height)

    def update_time_label(self, index, start_time):
        elapsed_time = time.time() - start_time
        self.canvas.itemconfig(self.labels[index], text=f"Time elapsed: {elapsed_time:.2f} s")

    def start_sorting_algorithms(self):
        for component in self.sorting_components:
            Thread(target=component.run).start()

    def highlight_bar(self, index, bar_index):
        self.canvas.itemconfig(self.bars[index][bar_index], fill="red")
        self.after(DELAY, lambda: self.canvas.itemconfig(self.bars[index][bar_index], fill="white"))

if __name__ == "__main__":
    app = ArrayVisualizer()
    app.mainloop()
```

```python
import random
from tkinter import * 

class Particle:
    def __init__(self, canvas, color, x, y):
        self.canvas = canvas
        self.x = x
        self.y = y
        self.energy = 1.0
        self.circle_id = self.canvas.create_oval(x - 5, y - 5, x + 5, y + 5, fill=color, outline="black", width=2)

    def update(self):
        dx = random.gauss(0, 2) * 2
        dy = random.gauss(0, 2) * 2
        self.x += dx
        self.y += dy
        self.energy -= 0.1
        self.canvas.move(self.circle_id, dx, dy)
        self.canvas.itemconfig(self.circle_id, fill=self.get_color_for_energy(self.energy))

    def get_color_for_energy(self, energy):
        if energy > 0.7:
            return 'red'
        elif energy > 0.3:
            return 'orange'
        elif energy > 0.1:
            return 'yellow'
        else:
            return 'blue'


class BrownianMotion:
    def __init__(self, width, height, num_particles, step_size, max_steps):
        self.width = width
        self.height = height
        self.num_particles = num_particles
        self.step_size = step_size
        self.max_steps = max_steps
        self.root = Tk()
        self.root.wm_title("Brownian Motion")
        self.canvas = Canvas(self.root, width=width, height=height)
        self.canvas.pack(padx=10, pady=10)
        self.particles = []

        self.create_particles()
        self.animate()

    def create_particles(self):
        for i in range(self.num_particles):
            x = random.uniform(0, self.width)
            y = random.uniform(0, self.height)
            particle = Particle(self.canvas, color='blue', x=x, y=y)
            self.particles.append(particle)

    def update_particles(self):
        for particle in self.particles:
            particle.update()

    def animate(self):
        self.update_particles()
        if self.max_steps:
            self.max_steps -= 1
            if self.max_steps <= 0:
                return
        self.canvas.after(100, lambda: self.animate())

if __name__ == "__main__":
    bm = BrownianMotion(width=800, height=600, num_particles=50, step_size=10, max_steps=1000)
    bm.root.mainloop()
```
