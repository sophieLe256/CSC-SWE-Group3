```python
import os

def contains_target_files(directory, extensions):
    """
    Check if a directory contains files with the given extensions
    """
    for item in os.listdir(directory):
        if os.path.isdir(os.path.join(directory, item)):
            # Recursively check in subdirectories
            if contains_target_files(os.path.join(directory, item), extensions):
                return True
        elif any(item.endswith(ext) for ext in extensions):
            return True
    return False

def print_directory_hierarchy(path, indent=0, extensions=('.py', '.html','.css','.png')):
    # Check if the path is a directory
    if os.path.isdir(path):
        # Only proceed if the directory contains target files
        if contains_target_files(path, extensions):
            # Print the directory name with a trailing slash
            print('    ' * indent + os.path.basename(path) + '/')
            # Iterate over all items in the directory
            for item in os.listdir(path):
                # Recursively call this function for each item
                print_directory_hierarchy(os.path.join(path, item), indent + 1, extensions)
    else:
        # If it's a file, print its name only if it has the desired extension
        if path.endswith(extensions):
            print('    ' * indent + os.path.basename(path))

# Example usage
print_directory_hierarchy('/Users/mm/Desktop/courier_tracking')



```

```python
import os

# Define the root directory
root_directory = '/Users/mm/Downloads/CSC-SWE-Group3-home-page-2'

# Function to print contents of all files
def print_file_contents(directory):
    for root, dirs, files in os.walk(directory):
        for file in files:
            file_path = os.path.join(root, file)
            if os.path.exists(file_path):
                try:
                    with open(file_path, 'rb') as f:  # Open in binary mode
                        content = f.read().decode('utf-8')  # Decode using utf-8
                        print(f'\n{file_path}:')  # Print the full file path
                        print(content)
                        print('=' * 30)
                except UnicodeDecodeError:
                    print(f"Error decoding file: {file_path}")
            else:
                print(f"File not found: {file_path}")

# Call the function with the root directory
print_file_contents(root_directory)
```

```python
import os
import graphviz

def visualize_directory_graphviz(directory_path):
    """Generates a directory tree visualization using Graphviz.

    Args:
        directory_path: The path to the directory to visualize.
    """

    dot = graphviz.Digraph(comment='Directory Structure')

    def add_nodes(path, parent_node):
        for item in os.listdir(path):
            full_path = os.path.join(path, item)            
            if os.path.isdir(full_path):
                node_name = item  
                dot.node(node_name, item)  # Create a node for the directory
                dot.edge(parent_node, node_name)  # Link to parent
                add_nodes(full_path, node_name)  # Recurse into subdirectories
            else:
                dot.node(item, item)  # Create a node for a file
                dot.edge(parent_node, item)

    add_nodes(directory_path, '')  # Start recursion from the root directory

    dot.render(view=True)  # Render and view the graph (requires Graphviz installed)

# Example usage:
visualize_directory_graphviz('/Users/mm/Desktop/courier_tracking')
```

```python
import os

def print_file_contents(directory, file_extensions):
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith(tuple(file_extensions)):
                rel_path = os.path.relpath(os.path.join(root, file), directory)
                print(f"{rel_path}:")
                with open(os.path.join(root, file), 'r', encoding='utf-8') as f:
                    content = f.read()
                    print(content)
                    print("\n================")
            elif any(file.endswith(ext) for ext in ['.jpg', '.png', '.gif', '.bmp', '.svg']):
                rel_path = os.path.relpath(os.path.join(root, file), directory)
                parent_dirs = os.path.split(rel_path)[0].split(os.sep)
                if len(parent_dirs) <= 2:
                    print(f"{'/'.join(parent_dirs)}:")
                else:
                    print(f"{'/'.join(parent_dirs[-2:])}:")
                print(f"Image: {file}")
                print("================")

if __name__ == "__main__":
    directory_path = input("Enter the directory path: ")
    file_extensions = ['.html', '.css', '.js']
    print_file_contents(directory_path, file_extensions)
```
