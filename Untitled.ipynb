```python
import os

def contains_target_files(directory, extensions):
    """
    Check if a directory contains files with the given extensions
    """
    for item in os.listdir(directory):
        if os.path.isdir(os.path.join(directory, item)):
            # Recursively check in subdirectories
            if contains_target_files(os.path.join(directory, item), extensions):
                return True
        elif any(item.endswith(ext) for ext in extensions):
            return True
    return False

def print_directory_hierarchy(path, indent=0, extensions=('.py', '.html')):
    # Check if the path is a directory
    if os.path.isdir(path):
        # Only proceed if the directory contains target files
        if contains_target_files(path, extensions):
            # Print the directory name with a trailing slash
            print('    ' * indent + os.path.basename(path) + '/')
            # Iterate over all items in the directory
            for item in os.listdir(path):
                # Recursively call this function for each item
                print_directory_hierarchy(os.path.join(path, item), indent + 1, extensions)
    else:
        # If it's a file, print its name only if it has the desired extension
        if path.endswith(extensions):
            print('    ' * indent + os.path.basename(path))

# Example usage
print_directory_hierarchy('/Users/mm/Desktop/courier_tracking')



```

```python
import os
import graphviz

def visualize_directory_graphviz(directory_path):
    """Generates a directory tree visualization using Graphviz.

    Args:
        directory_path: The path to the directory to visualize.
    """

    dot = graphviz.Digraph(comment='Directory Structure')

    def add_nodes(path, parent_node):
        for item in os.listdir(path):
            full_path = os.path.join(path, item)            
            if os.path.isdir(full_path):
                node_name = item  
                dot.node(node_name, item)  # Create a node for the directory
                dot.edge(parent_node, node_name)  # Link to parent
                add_nodes(full_path, node_name)  # Recurse into subdirectories
            else:
                dot.node(item, item)  # Create a node for a file
                dot.edge(parent_node, item)

    add_nodes(directory_path, '')  # Start recursion from the root directory

    dot.render(view=True)  # Render and view the graph (requires Graphviz installed)

# Example usage:
visualize_directory_graphviz('/Users/mm/Desktop/courier_tracking')
```

```python
import ast
import matplotlib.pyplot as plt

def generate_tree(filename):
    with open(filename) as file:
        code = file.read()

    tree = ast.parse(code)

    classes = [node for node in tree.body if isinstance(node, ast.ClassDef)]
    functions = [node for node in tree.body if isinstance(node, ast.FunctionDef)]
    variables = [node for node in tree.body if isinstance(node, ast.Assign)]

    # Create a figure with a 1x1 grid
    fig, ax = plt.subplots(1, 1)

    # Create a root node and add it to the graph
    root = ax.add_patch(plt.Circle((0.5, 0.5), 0.25, color="lightgray"))
    ax.text(0.5, 0.5, filename, ha="center", va="center")

    # Add nodes for each class
    for node in classes:
        ax.add_patch(plt.Circle((0.5, 0.5 - 0.25 * len(classes)), 0.25, color="lightgreen"))
        ax.text(0.5, 0.5 - 0.25 * len(classes), node.name, ha="center", va="center")

    # Add nodes for each function
    for node in functions:
        ax.add_patch(plt.Circle((0.5, 0.5 - 0.25 * len(functions)), 0.25, color="lightblue"))
        ax.text(0.5, 0.5 - 0.25 * len(functions), node.name, ha="center", va="center")

    # Add nodes for each variable
    for node in variables:
        ax.add_patch(plt.Circle((0.5, 0.5 - 0.25 * len(variables)), 0.25, color="lightyellow"))
        ax.text(0.5, 0.5 - 0.25 * len(variables), node.targets[0].id, ha="center", va="center")

    # Draw edges between nodes
    for i, node in enumerate(classes):
        for j, child in enumerate(node.body):
            if isinstance(child, ast.FunctionDef):
                ax.plot([0.5, 0.5 - 0.25 * len(classes)], [0.5 - 0.25 * len(functions), 0.5 - 0.25 * len(functions)], color="lightgreen")

    for i, node in enumerate(functions):
        for j, child in enumerate(node.body):
            if isinstance(child, ast.Assign):
                for target in child.targets:
                    if isinstance(target, ast.Name):
                        ax.plot([0.5, 0.5 - 0.25 * len(functions)], [0.5 - 0.25 * len(variables), 0.5 - 0.25 * len(variables)], color="lightblue")

    # Show the plot
    plt.show()

# Test the function
generate_tree(filepath)
```
